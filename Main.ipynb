```python
import os

mfp_user=os.environ['MFP_USER']
```

```python
import myfitnesspal

client = myfitnesspal.Client(mfp_user)
```

```python
days_per_month = [ 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
food_rows = []
```

```python
# Units: calories:kcal, carbs:g, fat:g, protein:g, sodium:mg, sugar:g
food_index = ['date', 'meal', 'name', 'amount', 'calories', 'carbohydrates', 'fat', 'protein', 'sodium', 'sugar']
```

```python
import datetime

def append_day (rows, year, month, day):
    daily_date = datetime.date(year, month, day)
    daily_record = client.get_date(year, month, day)
    daily_meals = daily_record.meals
    for i in range(0, len(daily_meals)):
        meal_entries = daily_meals[i]
        for j in range (0, len(meal_entries)):
            rows.append([ 
                daily_date, 
                daily_meals[i].name, 
                meal_entries[j].name.split(',', 1)[0],
                meal_entries[j].name.split(',', 1)[1],
                meal_entries[j].totals['calories'],
                meal_entries[j].totals['carbohydrates'],
                meal_entries[j].totals['fat'],
                meal_entries[j].totals['protein'],
                meal_entries[j].totals['sodium'],
                meal_entries[j].totals['sugar']
            ])

def append_month (rows, year, month):
    for i in range (1, days_per_month[month-1]+1):
        print('Processing '+str(year)+'-'+str(month)+'-'+str(i))
        append_day(rows, year, month, i)
```

```python
append_month(food_rows, 2019,12)
len(food_rows)
```

```python
import pandas as pd

df = pd.DataFrame(food_rows, columns = food_index)
df
```

```python
df.sort_values(by='fat', ascending=False).head(20)
```

```python
# Highest Protein per Calorie

df['carbs_per_cals']=df['carbohydrates']/df['calories']
df.sort_values(by='carbs_per_cals', ascending=False).head(20)
```

```python
# Highest Carbs per Calorie
df['prot_per_cals']=df['protein']/df['calories']
df.sort_values(by='prot_per_cals', ascending=False).head(20)
```
